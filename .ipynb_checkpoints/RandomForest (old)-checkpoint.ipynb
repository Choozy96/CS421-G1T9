```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
df = pd.read_csv('hotel_bookings.csv')
```

```python
df.info()
```

```python
df.head(50)
```

```python
df.isna().sum()
```

```python
#combining columns to form date

df["date"]=df["arrival_date_year"].astype(str) + "/" + df["arrival_date_month"] + "/"+ df["arrival_date_day_of_month"].astype(str)
```

```python
#convert date column to date type

df['date'] = pd.to_datetime(df['date'])
```

```python
#columns to drop 

to_drop=['arrival_date_year','arrival_date_month','arrival_date_week_number','arrival_date_day_of_month','reservation_status_date', 'reservation_status']

for column in df.columns:
    if column in to_drop:
        df=df.drop(column,axis=1)
```

```python
#drop null rows for column "Country"

df.dropna(subset=['country','children'],inplace=True)
```

```python
#columns to encode

to_encode = ["deposit_type", "customer_type", "hotel" , "meal", 'country','market_segment',"distribution_channel",'assigned_room_type', 'reserved_room_type']

for column in df.columns:
    if column in to_encode:
        le = preprocessing.LabelEncoder()
        le.fit(df[column].astype(str))
        df[column]=le.transform(df[column])
        keys = le.classes_
        values = le.transform(le.classes_)
        dictionary = dict(zip(keys, values))
        print(dictionary)
```

```python
df.info()
```

```python
df.head()
```

```python
#encoding both agent and company code

df['agent_encoded'] = df['agent'].notnull()
df['company_encoded'] = df['company'].notnull()
df["agent_encoded"] = df["agent_encoded"].astype(int)
df["company_encoded"] = df["company_encoded"].astype(int)
df=df.drop(['agent','company'],axis = 1)
```

```python
df.info()
```

```python
column_list=list(df.columns)
column_list.remove('is_canceled')
```

```python
# features
X = df[column_list]

# target
y = df['is_canceled']
```

### Train Test Split without Scaling and PCA

```python
X=X.drop(['date'],axis = 1)
# drop for now, random forest got trouble dealing with datetime format. need convert using toordinal i think
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
# split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=420)
```

```python
X.info()
```

```python
X.describe()
```

# Basic Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, confusion_matrix, auc
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]
print('Accuracy Score: {}\n\nConfusion Matrix:\n {}'.format(accuracy_score(y_test,y_pred), confusion_matrix(y_test,y_pred)))
```

```python
def ROC(y_test, y_prob):
    
    false_positive_rate, true_positive_rate, threshold = roc_curve(y_test, y_prob)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    
    plt.figure(figsize = (10,10))
    plt.title('Receiver Operating Characteristic')
    plt.plot(false_positive_rate, true_positive_rate, color = 'red', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1], linestyle = '--')
    plt.axis('tight')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
```

```python
ROC(y_test, y_prob)
```

### PCA

```python
X_pca = X_scaled.copy()
```

```python
X_pca.info()
```

```python
from sklearn.decomposition import PCA

# X_pca.drop('date', axis=1, inplace=True)

for i in range(1,24):
    pca = PCA(n_components=i, random_state=420)
    components = pca.fit_transform(X_pca)
    print(i)
    print('explained var ratio', pca.explained_variance_ratio_)
    print('sum', sum(pca.explained_variance_ratio_))
    print()
```

```python
# 11 randomly chosen, might want to change number of principle components
n = 11 
pca = PCA(n_components=n, random_state=69)
pca_components = pca.fit_transform(X_pca)
print('explained var ratio', pca.explained_variance_ratio_)
print('sum', sum(pca.explained_variance_ratio_))

columns = []
for i in range(n):
    x = "pc" + str(i)
    columns.append(x)

# convert numpy array to dataframe
X_pca = pd.DataFrame(pca_components, columns=columns)
X_pca
```

```python
# Train Test Split with Scaling and PCA
X_pca_scaled_train, X_pca_scaled_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=420)
```

```python
model2 = RandomForestClassifier()
model2.fit(X_pca_scaled_train, y_train)
```

```python
y_pred = model2.predict(X_pca_scaled_test)
y_prob = model2.predict_proba(X_pca_scaled_test)[:,1]
print('Accuracy Score: {}\n\nConfusion Matrix:\n {}'.format(accuracy_score(y_test,y_pred), confusion_matrix(y_test,y_pred)))
```
